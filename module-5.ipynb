1. **Using Unicode Characters and Strings**
    - Write a program that takes a string containing unicode characters and counts the occurrences of each character.

```python
def count_unicode_characters(text):
    # Create an empty dictionary to store character counts
    char_counts = {}

    # Iterate over each character in the string
    for char in text:
        # Increment the count for the character in the dictionary
        char_counts[char] = char_counts.get(char, 0) + 1

    return char_counts

# Test the function
unicode_text = "Hello, 你好, स्वागत"
character_counts = count_unicode_characters(unicode_text)

# Print the character counts
for char, count in character_counts.items():
    print(f"{char}: {count}")
```

3. **Selecting and Processing Data**
    - Write a program that reads data from a CSV file, processes it (e.g., filters rows, calculates statistics), and saves the processed data to a new CSV file.

```python
import csv

def read_csv(input_file):
    data = []
    with open(input_file, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

def process_data(data):
    # Filter rows where the 'category' column is 'A'
    filtered_data = [row for row in data if row['category'] == 'A']
    
    # Calculate statistics (e.g., total sales for 'category A')
    total_sales = sum(float(row['sales']) for row in filtered_data)
    
    return filtered_data, total_sales

def write_csv(output_file, processed_data, total_sales):
    with open(output_file, 'w', newline='') as csvfile:
        fieldnames = processed_data[0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        writer.writeheader()
        for row in processed_data:
            writer.writerow(row)
        
        # Write total sales as a separate row
        writer.writerow({'category': 'Total Sales', 'sales': total_sales})

# Example usage:
input_file = 'sample_data.csv'
output_file = 'processed_data.csv'

# Read data from CSV file
data = read_csv(input_file)

# Process the data
processed_data, total_sales = process_data(data)

# Write processed data to a new CSV file
write_csv(output_file, processed_data, total_sales)
```

4. **Creating Email Data Visualizations**
    - Develop a program that reads email data (e.g., from a CSV file) and generates visualizations such as histograms of email frequencies by sender and time of day.

```python
import csv
import matplotlib.pyplot as plt
from collections import Counter

def read_email_data_csv(input_file):
    data = []
    with open(input_file, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

def visualize_email_frequency_by_sender(email_data):
    senders = [email['sender'] for email in email_data]
    sender_counts = Counter(senders)
    
    plt.bar(range(len(sender_counts)), list(sender_counts.values()), align='center')
    plt.xticks(range(len(sender_counts)), list(sender_counts.keys()), rotation='vertical')
    plt.xlabel('Sender')
    plt.ylabel('Frequency')
    plt.title('Email Frequency by Sender')
    plt.tight_layout()
    plt.show()

def visualize_email_frequency_by_time(email_data):
    timestamps = [email['timestamp'] for email in email_data]
    hours = [int(timestamp.split()[1].split(':')[0]) for timestamp in timestamps]
    hour_counts = Counter(hours)
    
    plt.bar(range(len(hour_counts)), list(hour_counts.values()), align='center')
    plt.xticks(range(len(hour_counts)), list(hour_counts.keys()))
    plt.xlabel('Hour of Day')
    plt.ylabel('Frequency')
    plt.title('Email Frequency by Time of Day')
    plt.tight_layout()
    plt.show()

# Example usage:
input_file = 'email_data.csv'

# Read email data from CSV file
email_data = read_email_data_csv(input_file)

# Visualize email frequency by sender
visualize_email_frequency_by_sender(email_data)

# Visualize email frequency by time of day
visualize_email_frequency_by_time(email_data)
```
